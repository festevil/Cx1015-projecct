```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
air=pd.read_csv('AviationData.csv', encoding='ISO-8859-1')
```

```python
air.dtypes
```

```python
air
```

```python
airfiltered=air[(air["Investigation.Type"]=="Accident")]
```

```python
airfiltered
```

```python
air1 = pd.DataFrame(airfiltered[['Aircraft.Damage','Weather.Condition']])
```

```python
air1=air1.dropna(how='any',axis=0)
```

```python
air1
```

```python
jointDF = pd.concat([air1['Aircraft.Damage'], air1['Weather.Condition']], axis = 1, join_axes = [air1['Aircraft.Damage'].index])
```

```python
sb.heatmap(jointDF.corr(), vmin = -1, vmax = 1, annot = True, fmt=".2f")
```
